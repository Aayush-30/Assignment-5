```python
import statistics
def visualize(filename):
    df = pd.read_csv(filepath_or_buffer = filename,header=None)
    means = []
    mins = []
    maxs = []
    for i in df.columns:
        means.append(statistics.mean(df[i]))
        mins.append(min(df[i]))
        maxs.append(max(df[i]))
    plt.plot(means,label='Average')
    plt.plot(maxs,label='Maximum')
    plt.plot(mins,label='Minimum')
    plt.title(filename)
    plt.ylabel("Values")
    plt.xlabel("Feature#")
    plt.legend()
    plt.show()
```

```python
def detect_problems(filename):
    df = pd.read_csv(filepath_or_buffer = filename,header=None)
    means = []
    mins = []
    maxs = []
    for i in df.columns:
        means.append(statistics.mean(df[i]))
        mins.append(min(df[i]))
        maxs.append(max(df[i]))
    if((maxs[0]==0) & (maxs[20]==20)):
            print("I doubt the integrity of this dataset")
    elif(sum(mins)==0):
            print("Why is this happening?")
    else:
        print("Dataset looks good")
```

```python
for j in dataset:
    print("File name:",j)
    visualize(j)
    detect_problems(j)
```
